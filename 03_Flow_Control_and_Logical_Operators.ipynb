In this exercise you need to request from user to input their weight and height, then calculate BMI and find the category person belongs.

```python
## Solution:

weight=float(input('Please enter your weight(in kg)?\n'))
height=float(input('Please enter your height(in m)?\n'))

bmi=weight/(height*height)
print(bmi)

if bmi<18.5:
  print('You are underwight')
elif bmi>=18.5 and bmi<25:
  print('You are normal weight')

elif bmi>=25 and bmi<30:
  print('You are overweight')
else:
  print('You are obese!!! Please go to the doctor!')
```

## Exercise #4 - Love Calculator

So this exercise is a little hard. So there is a calculator called Love Calculator which calculates the love score.

Instructions:

Take both people's names and check for the number of times the letters in the word TRUE occurs. Then check for the number of times the letters in the word LOVE occurs. Then combine these numbers to make a 2 digit number.

- For Love Scores less than 10 or greater than 90, the message should be:

"Your score is **x**, you go together like coke and mentos."


- For Love Scores between 40 and 50, the message should be:

"Your score is **y**, you are alright together."


- Otherwise, the message will just be their score. e.g.:

"Your score is **z**.

!!! Explain count and lower functions

```python
## Solution
your_name=input('What is your name?\n')
their_name=input('What is his/her name?\n')

#combining strings and lowercasing
combined_name=your_name+their_name
combined_name_lowercase=combined_name.lower()

#calculating love score
t=combined_name_lowercase.count('t')
r=combined_name_lowercase.count('r')
u=combined_name_lowercase.count('u')
e=combined_name_lowercase.count('e')

l=combined_name_lowercase.count('l')
o=combined_name_lowercase.count('o')
v=combined_name_lowercase.count('v')
e=combined_name_lowercase.count('e')

love_score=(t+r+u+e)*10+(l+o+v+e)


#conditions
if love_score<10 and love_score>90:
  print(f'Your score is {love_score}, you go together like coke and mentos')
elif love_score>=40 and love_score<=50:
  print(f'Your love score is {love_score}. you are alright together')
else:
  print(f'Your score is {love_score}')
```
